```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np




# File to Load (Remember to change these)
cityDataToLoad = "data/city_data.csv"
rideDataToLoad = "data/ride_data.csv"

# Read the City and Ride Data
cityData = pd.read_csv(cityDataToLoad, encoding="utf-8")
rideData = pd.read_csv(rideDataToLoad, encoding="utf-8")

# Combine the data into a single dataset
cityRideData = cityData.merge(rideData, on="city" ,left_index=True , how="inner")
# Dropping the duplicate city rows keeping the first row with lowest index
#cityRideData = cityRideData.drop_duplicates(subset="city",keep="first")
# Sorting the DataFrame by index key
cityRideData = cityRideData.sort_index()
# Rearranging the columns
cityRideData = cityRideData[["city","date","fare","ride_id","driver_count","type"]]
cityRideData.head()

```

```python
# Defining colors for Pyber charts
colorScheme = {'Gold':'#FFD700', 'Sky Blue':'#87CEEB', 'Coral':'#EE6A50'}
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Extracting the City Types
cityTypes = cityRideData['type'].unique()

pltData = {}
# Calculating Average Fare ($) Per City
averageFarePerCity = cityRideData.groupby(['type','city']).mean()
averageFarePerCity = averageFarePerCity.reset_index()
# Calculating Total Number of Rides Per City
totalNumberOfRides = cityRideData.groupby(['type','city']).count()
totalNumberOfRides = totalNumberOfRides.reset_index()


# Calculating the Total Number of Drivers Per City
#totalNumberofDrivers = cityRideData.groupby(['type','city']).count()
#totalNumberofDrivers = totalNumberofDrivers.reset_index()


for cityType in cityTypes:
    key = f'x{cityType.capitalize()}'
    key1 = f'y{cityType.capitalize()}'
    key2 = f'z{cityType.capitalize()}'
    pltData[key] = totalNumberOfRides['ride_id'][totalNumberOfRides['type'] == cityType]
    pltData[key1]= averageFarePerCity['fare'][averageFarePerCity['type'] == cityType]
    pltData[key2] = cityData['driver_count'][cityData['type'] == cityType]
    #pltData[key2] = totalNumberofDrivers['driver_count'][cityData['type'] == cityType]


# Build the scatter plots for each city types
# Incorporate the other graph properties

fig = plt.figure()
ax = fig.add_axes([0,0,0.8,0.8])

ax.scatter(pltData['xUrban'],pltData['yUrban'],s=pltData['zUrban']*10,alpha=0.7,facecolor=colorScheme['Coral'],edgecolor="black",lw=1)
ax.scatter(pltData['xSuburban'],pltData['ySuburban'],s=pltData['zSuburban']*10,alpha=0.7,facecolor=colorScheme['Sky Blue'],edgecolor="black",lw=1)
ax.scatter(pltData['xRural'],pltData['yRural'],s=pltData['zRural']*10,alpha=0.7,facecolor=colorScheme['Gold'],edgecolor="black",lw=1)
ax.grid()
ax.set_xlabel("Total Number of Rides (Per City)" ,fontsize="large")
ax.set_ylabel("Average Fare ($)", fontsize="large")
ax.set_title("Pyber Ride Sharing Data (2016)")

# Create a legend
lgnd = ax.legend(cityTypes,fontsize="medium",title="City Types",loc="best",scatterpoints=1,labelspacing=0.5)
lgnd.set_title("City Types", prop = {'size':'large'})
for handle in lgnd.legendHandles:
    handle.set_sizes([25.0])

# Incorporate a text label regarding circle size
left = 1.03
top = 0.7
ax.text(left, top, 'Note:',
        fontsize = "medium",
        horizontalalignment='left',
        verticalalignment='center',
        rotation='horizontal',
        transform=ax.transAxes)
ax.text(1.03,0.65,'Circle size correlates with driver count per city.',
        fontsize = "medium",
        horizontalalignment='left',
        verticalalignment='center',
        rotation='horizontal',
        transform=ax.transAxes)
plt.tight_layout
# Save Figure
plt.savefig("data/PyberBubbleChart.png")
# Show plot
plt.show()
```

```python
# Defining explode values for various pie charts in the order Rural, Suburban and Urban
explodeValues = (0,0,0.2)
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Calculating the total fare for each city type
totalFareCityType = cityRideData.groupby(['type']).sum()
totalFareCityType = totalFareCityType[['fare']]


# Build Pie Chart
fig, ax = plt.subplots(1,1,figsize=(10,6))
#explFare = (0,0,0.2)
fares = totalFareCityType['fare']
labelFare = totalFareCityType.index
pieChartFare = ax.pie(fares,
                   colors = colorScheme.values(), 
                   explode = explodeValues, 
                   labels = labelFare, 
                   shadow = True ,
                   wedgeprops= {'linewidth': 3},
                   startangle = 150,
                   autopct="%1.1f%%")
ax.set_title("% of Total Fares by City Type",pad = 0.5)

# Save Figure
plt.savefig("data/TotalFaresPie.png")
# show figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Calculating the total rides for each city type
totalRidesByCityType = cityRideData.groupby(['type']).count()
totalRidesByCityType = totalRidesByCityType[['ride_id']]


fig, ax = plt.subplots(1,1,figsize=(10,6))
# explRides = (0,0,0.2)
rides = totalRidesByCityType['ride_id']
labelRides = totalRidesByCityType.index
pieChartRides = ax.pie(rides,
                   colors = colorScheme.values(), 
                   explode = explodeValues, 
                   labels = labelRides, 
                   shadow = True ,
                   wedgeprops= {'linewidth': 3},
                   startangle = 140,
                   autopct="%1.1f%%")
ax.set_title("% of Total Rides by City Type")

# Save Figure
plt.savefig("data/TotalRidesPie.png")
# show figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# Calculating the total drivers for each city type
totalDriversByCityType = cityData.groupby(['type']).sum()
totalDriversByCityType = totalDriversByCityType[['driver_count']]


fig, ax = plt.subplots(1,1,figsize=(10,6))
#explDrivers = (0,0,0.2)
drivers = totalDriversByCityType['driver_count']
labelDrivers = totalDriversByCityType.index
pieChartDrivers = ax.pie(drivers,
                   colors = colorScheme.values(), 
                   explode = explodeValues, 
                   labels = labelDrivers, 
                   shadow = True ,
                   wedgeprops = {'linewidth': .5, 'edgecolor': 'black'},
                   startangle = 170,
                   counterclock = True,
                   autopct="%1.1f%%")
ax.set_title("% of Total Drivers by City Type")


# Save Figure
plt.savefig("data/TotalDriversPie.png")
# show figure
plt.show()
```
