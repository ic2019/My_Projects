# Report

## Player Count

* Display the total number of players

```python
uniquePlayersData = purchase_data.drop_duplicates(['SN'])
totalPlayers = len(uniquePlayersData)
# Creating the Summary Data Frame for Total Players
totalPlayerSummary = pd.DataFrame({'Total Players' : [totalPlayers]})
# Formatting the Data Frame
totalPlayerSummary = totalPlayerSummary.applymap(lambda x: f'{x:,}')
# Displaying the Summary Data
display(totalPlayerSummary)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame
-----------------

```python
# Calculating Total number of Unique items.
totalUniqueItems = purchase_data['Item ID'].nunique()

# Calculating Average Purchase Price.
avg_purchase_price = purchase_data['Price'].mean()

# Calculating Total Number of Purchases
total_purchases = len(purchase_data)

# Calculating Total Revenue.
total_revenue = avg_purchase_price * total_purchases

# Creating and Displaying Summary Data
purchase_summary = pd.DataFrame({'Number of Unique Items' : [totalUniqueItems],
                                'Average Price' : [avg_purchase_price],
                                'Number of Purchases' : [total_purchases],
                                'Total Revenue': [total_revenue]})
# Formatting the Summary Data
purchase_summary[['Average Price','Total Revenue']] = purchase_summary[['Average Price','Total Revenue']].applymap(lambda x: f'${x:,.2f}')
# Displaying the summary data
display(purchase_summary)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed
---------------

```python
# Calculating Percentage and Count of Male Players
maleCount = uniquePlayersData['Gender'].value_counts()['Male']
malePercent = (maleCount / totalPlayers)*100

# Calculating Percentage and Count of Female Players

femaleCount = uniquePlayersData['Gender'].value_counts()['Female']
femalePercent = (femaleCount / totalPlayers)*100

# Calculating Percentage and Count of Other Gender types

othersCount = totalPlayers - (maleCount + femaleCount)
othersPercent = (othersCount / totalPlayers)*100

# Setting the index column
genderTypes = uniquePlayersData['Gender'].value_counts().index

# Creating the Data Frame
genderSummary = pd.DataFrame({'Total': [maleCount , femaleCount , othersCount],
                             'Percentage of Players': [malePercent , femalePercent,othersPercent]},
                            index=[genderTypes])


# Formatting the Gender Summary Data
genderSummary[['Percentage of Players']] = genderSummary[['Percentage of Players']].applymap(lambda x: f'{x:.2f}%')


# Displaying the Gender Summary data 
display(genderSummary)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculating Total Purchase Count
totalPurchaseCount = purchase_data['Gender'].count()
# Calculating purchase count by different gender types.
maleCountTot = purchase_data['Gender'].value_counts()['Male']
femaleCountTot = purchase_data['Gender'].value_counts()['Female']
othersCountTot = purchase_data['Gender'].value_counts()['Other / Non-Disclosed']
index_values = purchase_data['Gender'].value_counts().index

# Calculating Average Purchase price by gender type.
avgPurchasePriceMale = purchase_data.loc[(purchase_data['Gender'] == 'Male')]['Price'].sum() / maleCountTot
avgPurchasePriceFemale = purchase_data.loc[(purchase_data['Gender'] == 'Female')]['Price'].sum() / femaleCountTot
avgPurchasePriceOthers = purchase_data.loc[(purchase_data['Gender'] == 'Other / Non-Disclosed')]['Price'].sum() / othersCountTot

# Calculating Total Purchase Value
totalPurchaseValueMale = maleCountTot * avgPurchasePriceMale
totalPurchaseValueFemale = femaleCountTot * avgPurchasePriceFemale
totalPurchaseValueOthers = othersCountTot * avgPurchasePriceOthers

# Calculating Average Purchase by Person based on gender type.
avgPurchaseByPersonMale = totalPurchaseValueMale / maleCount
avgPurchaseByPersonFemale = totalPurchaseValueFemale / femaleCount
avgPurchaseByPersonOthers = totalPurchaseValueOthers / othersCount

# Creating a Summary Table.

summaryByGenderTypePurchase = pd.DataFrame({'Purchase Count': [maleCountTot, femaleCountTot, othersCountTot],
                                           'Average Purchase Price': [avgPurchasePriceMale, avgPurchasePriceFemale, avgPurchasePriceOthers],
                                            'Total Purchase Value': [totalPurchaseValueMale, totalPurchaseValueFemale, totalPurchaseValueOthers],
                                            'Average Total Purchase by Person': [avgPurchaseByPersonMale, avgPurchaseByPersonFemale,avgPurchaseByPersonOthers]},index=[index_values])

# Formatting the data
summaryByGenderTypePurchase.loc[:,['Average Purchase Price','Total Purchase Value','Average Total Purchase by Person']] = summaryByGenderTypePurchase[['Average Purchase Price','Total Purchase Value','Average Total Purchase by Person']].applymap(lambda x: f'${x:,.2f}')



display(summaryByGenderTypePurchase)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame
------------------

* Creating Bins for purchase_data by age and categorzing the purchase data into the bins.

```python
# Creating Bins for purchase_data by age and categorzing the purchase data into the bins.
# Creating Bins for various Age Groups
ageBins = [ 0,9,14,19,24,29,34,39,48]

# Creating Bin Labels
ageLabels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Categorizing the purchase data into various Bins
purchase_data['ageGroup'] = pd.cut(purchase_data['Age'], bins = ageBins , labels=ageLabels)

# calculating purchase counts by age group
countPurchaseDataByAge = purchase_data['ageGroup'].value_counts()

# Calculating purchase count
lessthan10 = countPurchaseDataByAge['<10']
tenTo14 = countPurchaseDataByAge['10-14']
fifteenTo19 = countPurchaseDataByAge['15-19']
twentyTo24 = countPurchaseDataByAge['20-24']
twentyfiveTo29 = countPurchaseDataByAge['25-29']
thirtyTo34 = countPurchaseDataByAge['30-34']
thirtyfiveTo39 = countPurchaseDataByAge['35-39']
fourtyPlus = countPurchaseDataByAge['40+']

# Calculate Average Purchase price by age groups.

lessthan10AvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '<10'].mean()
tenTo14AvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '10-14'].mean()
fifteenTo19AvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '15-19'].mean()
twentyTo24AvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '20-24'].mean()
twentyfiveTo29AvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '25-29'].mean()
thirtyTo34AvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '30-34'].mean()
thirtyfiveTo39AvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '35-39'].mean()
fourtyPlusAvgPrice = purchase_data['Price'][purchase_data['ageGroup'] == '40+'].mean()

# Calculating Total Purchase Value by Age Group
totalLess10 = lessthan10 * lessthan10AvgPrice
total10To14 = tenTo14 * tenTo14AvgPrice
total15To19 = fifteenTo19 * fifteenTo19AvgPrice
total20To24 = twentyTo24 * twentyTo24AvgPrice
total25To29 = twentyfiveTo29 * twentyfiveTo29AvgPrice
total30To34 = thirtyTo34 * thirtyTo34AvgPrice
total35to39 = thirtyfiveTo39 * thirtyfiveTo39AvgPrice
total40Plus =  fourtyPlus * fourtyPlusAvgPrice

# Calculating Average Total Purchase Per Person
avgPersonLess10 = totalLess10 / lessThanTen
avgPerson10To14 = total10To14 / tenTofourteen
avgPerson15To19 = total15To19 / fifteenToninteen
avgPerson20To24 = total20To24 / twentTotwntyfour
avgPerson25To29 = total25To29 / twentyfiveTotwentynine
avgPerson30To34 = total30To34 / thirtyTothirtyfour
avgPerson35To39 = total35to39 / thirtyfiveTothirtynine
avgPerson40Plus = total40Plus / fourtyPlus

# Creating the Summary Dataframe for Purchase Data by Age Group.
purchaseSummaryByAge = pd.DataFrame({'Purchase Count': [lessthan10,tenTo14, fifteenTo19, twentyTo24, twentyfiveTo29, thirtyTo34, thirtyfiveTo39, fourtyPlus],
                                     'Average Purchase Price': [lessthan10AvgPrice, tenTo14AvgPrice, fifteenTo19AvgPrice, twentyTo24AvgPrice, twentyfiveTo29AvgPrice, thirtyTo34AvgPrice, thirtyfiveTo39AvgPrice, fourtyPlusAvgPrice ],
                                     'Total Purchase Value': [totalLess10, total10To14, total15To19, total20To24, total25To29, total30To34, total35to39, total40Plus],
                                     'Avg Total Purchase per Person' : [avgPersonLess10, avgPerson10To14, avgPerson15To19, avgPerson20To24, avgPerson25To29, avgPerson30To34, avgPerson35To39, avgPerson40Plus]}, index = ageLabels)

# Displaying the Summary Purchase Data by Age Group after formatting.
purchaseSummaryByAge[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']] = purchaseSummaryByAge[['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']].applymap(lambda x: f'${x:,.2f}')

# Displaying the Purchase summary by Age Group.
display(purchaseSummaryByAge)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame
--------------------

```python
# Performing calculations and extracting the data
# Extracting the count of all spenders into a dataframe
topSpenders1 = pd.DataFrame(purchase_data['SN'].value_counts())
topSpenders1 = topSpenders1.reset_index()
topSpenders1.columns = ['SN','Purchase Count']

#Extracting the Average Purchase Price for each Spender into another Dataframe.
topSpenders2 = purchase_data.groupby(['SN']).mean()
topSpenders2 = topSpenders2.reset_index()[['SN','Price']]

# Merging both Data Frames on Spender Name ['SN']
topSpenders = topSpenders1.merge(topSpenders2,on=['SN'],left_index=False,right_index=False)

# Calculating Total Purchase Value
topSpenders['Total Purchase Value'] = topSpenders['Purchase Count'] * topSpenders['Price']

# Sorting the Data in Descending order of Total Purchase Value
topSpenders.sort_values(by='Total Purchase Value', ascending=False,inplace=True)

# Formatting the Price columns with currency and 2 decimal places
topSpenders[['Price','Total Purchase Value']] = topSpenders[['Price','Total Purchase Value']].applymap(lambda x: f'${x: ,.2f}')

# Saving only the top 5 Spenders
topSpenders = topSpenders.head()

# Renaming column names
topSpenders.rename(columns={'Price':'Average Purchase Price'},inplace=True)

# Setting the index to Spender Name
topSpenders.set_index(['SN'],inplace=True)
display(topSpenders)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame
-------------

```python
# Retrieve the Item ID, Item Name, and Item Price columns
popularDataRaw = purchase_data.loc[:,['Item ID', 'Item Name', 'Price']]
```

```python
# Group the data by Item ID and Item Name and renaming the column with meaningful name
popularGrouped = popularDataRaw.groupby(['Item ID','Item Name']).count()

popularGrouped.rename(columns={'Price' : 'Purchase Count'},inplace=True)

# Extracting Price
popularPriceData =  purchase_data.drop_duplicates(['Item ID', 'Item Name'])[['Item ID', 'Item Name', 'Price']]

popularGrouped = popularGrouped.merge(popularPriceData,on=['Item ID','Item Name'],how = 'left')

# Calculating Total Purchase Value
purchaseValue = popularGrouped['Purchase Count'] * popularGrouped['Price']
popularGrouped['Total Purchase Value'] = purchaseValue

# Renaming the Price column
popularGrouped.rename(columns={'Price':'Item Price'},inplace=True)

# Sorting the data 
popularList = popularGrouped.sort_values(by=['Purchase Count'],ascending=False)

popularList.set_index(['Item ID','Item Name'],inplace=True)
# Formatting the Data
popularList[['Item Price','Total Purchase Value']] = popularList[['Item Price','Total Purchase Value']].applymap(lambda x:f'${x: ,.2f}')
# Printing a Preview of Popular Play List.
display(popularList.head())
```

```python
# Sorting the Popular Player List By Descending order of Total Purchase Value.
topProfitableList = popularGrouped.sort_values(by=['Total Purchase Value'], ascending=False)

topProfitableList.set_index(['Item ID','Item Name'],inplace=True)

# Formatting the Data
topProfitableList[['Item Price','Total Purchase Value']] = topProfitableList[['Item Price','Total Purchase Value']].applymap(lambda x:f'${x: ,.2f}')

# Display a Preview of the top Profitable
display(topProfitableList.head())
```
